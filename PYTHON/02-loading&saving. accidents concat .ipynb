```python
import pandas as pd
```

```python
file_path = 'data/01_ ACCIDENTES POR TIPO EN  DISTRITOS.xls'
all_accidents = pd.read_excel(file_path, index_col=0, header=7, sheet_name=None)
```

```python
accidents_2013 = pd.read_excel(file_path, index_col=0, header=4, sheet_name='2013')
accidents_2013['YEAR']='2013'
```

```python
excel_l = []
for year in all_accidents.keys():
    all_accidents[year]['YEAR']=year
for n in all_accidents: 
    if n == '2013':
        continue
    else:
        excel_l.append(all_accidents[n])
    
all_accidents_df = pd.concat(excel_l)
all_accidents_total = all_accidents_df.append(accidents_2013)

all_accidents_total
```
